```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
g = nx.erdos_renyi_graph(30, 0.2)
```

30 nodes, and each edge exists with probability 0.2

```python
nx.draw(g)
```

```python
g = nx.erdos_renyi_graph(30, 0.08)
nx.draw(g)
```

G(30, 0.05) contains many isolated nodes. 

```python
g = nx.erdos_renyi_graph(600, 0.02)
nx.draw(g)
```

```python
nx.diameter(g)
```

```python
N = 100
epsilon = -1

giant_component = nx.erdos_renyi_graph(100, (1 + epsilon) / N)

nx.draw(giant_component)
print('N = {}; epsilon = {}'.format(N, epsilon))
```
